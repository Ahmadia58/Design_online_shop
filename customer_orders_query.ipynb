```python
import pandas as pd
from sqlalchemy import create_engine

# ساختن engine با اطلاعات دیتابیس
engine = create_engine("postgresql+psycopg2://postgres:1013@localhost:5432/ecommerce_dwh")

# خواندن کوئری از فایل
with open("customer_orders.sql", "r", encoding="utf-8") as f:
    query = f.read()

# اجرای کوئری و گرفتن داده‌ها در DataFrame
df = pd.read_sql(query, engine)

# ذخیره خروجی به فایل Excel
df.to_excel("C:/Users/Ahmad/project8/customer_orders.xlsx", index=False)

print("✅ گزارش با موفقیت در فایل Excel ذخیره شد.")

```
